```python
# LETS REMOVE COLUMNS WITH ZERO STANDARD STandard deviation
def cols_with_zero_std(data):
    cols_with_zero_std=[]
    num_cols=[col for col in data.columns if data[col].dtype!='O']
    for i in num_cols:
        if data[i].std()==0:
            cols_with_zero_std.append(i)
    return cols_with_zero_std
```

```python
with_zero_std=cols_with_zero_std(wafer)
```

```python
# LETS REMOVE REDUNDANT COLUMNS(COLUMNS HAVING MORE MISSING VALUES THAN THE LIMIT)
def get_redundant_cols(data,missing_thrash):
    col_missing_ratio=data.isnull().sum().div(data.shape[0])
    cols_to_drop=list(col_missing_ratio[col_missing_ratio>missing_thrash].index)
    return cols_to_drop
```

```python
col_missing_ratio=wafer.isnull().sum().div(wafer.shape[0])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import KNNImputer
from sklearn.preprocessing import RobustScaler
```

```python
step1=KNNImputer(n_neighbors=3)
step2=RobustScaler()
preprocess_pipe=Pipeline([('step1',step1),('step2',step2)])
```

```python
X_trans=preprocess_pipe.fit_transform(X)
```

```python
X_trans
```

```python
# WE KNOW THAT TARGET COLUMN IS HIGHLY IMBALENCED, LETS HANDLE IT USING SMOTE
from imblearn.combine import SMOTETomek
resample=SMOTETomek(sampling_strategy='auto')
X_res,y_res=resample.fit_resample(X_trans,y)
```

```python
print("Before resampling, Shape of training instances: ", np.c_[X_trans,y].shape)
print("After resampling, Shape of training instances: ", np.c_[X_res,y_res].shape)
```

## LETS TRAIN THE MODEL 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.30, random_state=42)
```

```python
models={'KNN':KNeighborsClassifier(n_neighbors=5),'RandomForestClassifier':RandomForestClassifier(),'LogisticRegression':LogisticRegression()}
for i in range(len(models)):
    model=list(models.values())[i]
    model.fit(X_train,y_train)
    y_pred1=model.predict(X_test)
    print("ACCURACY SCORE OF ",list(models.keys())[i], " = ",accuracy_score(y_test,y_pred1))
```
